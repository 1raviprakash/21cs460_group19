```python
from statsmodels.tsa.stattools import adfuller
```

```python
def ad_test(dataset):
    dftest=adfuller(dataset,autolag='AIC')
    print("1.ADF :",dftest[0])
    print("2.P-value :",dftest[1])
    print("3.Num of Lag :",dftest[2])
    print("4.Num of observation used for ADF Regresstion And Critical value calculation:",dftest[3])
    print("5.Criitical values:")
    for key, val in dftest[4].item():
        print("\t",key,":",val)
```

```python
ad_test(df['T2M'])
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
SARIMAX?
```

```python
print(df.shape)
```

```python
train=df.iloc[:-30]
test=df.iloc[-30:]
print(train.shape,test.shape)
```

```python
# sarimax_model=SARIMAX(tempdata['Temp'],order=(1,1,1),seasonal_order=(1,1,1,4),exog=tempdata['Earth Skin Temperature (C)'])

model=SARIMAX(train['T2M'],order=(4,0,4),seasonal_order=(3,0,3,12),exog=train['PS'])
model=model.fit()
model.summary()
```

```python
start=len(train)
end=len(train)+len(test)-1
pred=model.predict(start=start,end=end,typ='levels',exog=test['PS'])
print(pred)
```

```python
pred.plot(legend=True)
test['T2M'].plot(legend=True)
```

```python
test['T2M'].mean()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse=sqrt(mean_squared_error(pred,test['T2M']))
print(rmse)
```
